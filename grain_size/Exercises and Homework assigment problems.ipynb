## Example 2.4

```python
import pandas as pd
from frake import GraphFunctions as GF
import matplotlib.pyplot as plt
```

```python
D=[4.75, 2, 0.85, 0.425, 0.25, 0.15, 0.106, 0.075, 0.075]
W=[12.5, 233.67, 144.13, 75.75, 25.47, 10.14, 5.5, 1.7, 1]
Wt=sum(W) #Total weight
Wt
```

```python
GSD={'d': D, 'W':W}
GSD=pd.DataFrame(data=GSD)
GSD['R']=1.0
GSD['P']=1.0
for i in range(len(GSD['d'])):
    if (i>0):
        GSD['R'][i]=GSD['R'][i-1]+(GSD['W'][i]*100/Wt)
    else:
        GSD['R'][i]=GSD['W'][i]*100/Wt
GSD['P']=100-GSD['R']
GSD=[GSD]
GSD
```

```python
lims=[[0.001, 100],[0,100],1,0,1 ]
GF.PlotAll(GSD, ['1'], 'd', 'P', legendMode=False, hold=1, xlabel=r'Particle diameter ($d$) [mm]',\
          ylabel=r'Percentage finer', xsize=5, ysize=3, lims=lims, linewith=2)
plt.xscale('log')
plt.grid(True)
plt.grid(b=True, which='minor', linestyle='--')
plt.savefig('GSD.svg')
```

## Example 2.5

```python
D=[4.75, 2, 0.85, 0.425, 0.25, 0.15, 0.106, 0.075, 0.075]
W=[0, 12.12, 25.36, 32.15, 33.14, 33.78, 47.56, 50.14, 100.25]
Wt=sum(W) #Total weight
GSD={'d': D, 'W':W}
GSD=pd.DataFrame(data=GSD)
GSD['R']=1.0
GSD['P']=1.0
for i in range(len(GSD['d'])):
    if (i>0):
        GSD['R'][i]=GSD['R'][i-1]+(GSD['W'][i]*100/Wt)
    else:
        GSD['R'][i]=GSD['W'][i]*100/Wt
GSD['P']=100-GSD['R']
GSD=[GSD]
GSD
```

```python
D=[0.066, 0.045, 0.036, 0.025, 0.015, 0.011, 0.007, 0.005, 0.004, 0.003, 0.0018, 0.0012]
P=[84.5, 74.3, 68.3, 58.2, 48.4, 42.3, 34.6, 28.1, 24.3, 20.1, 16.2, 12.3]
GSD1={'d': D, 'P':P}
GSD1=pd.DataFrame(data=GSD1)
GSD.append(GSD1)
```

```python
lims=[[0.001, 100],[0,100],1,0,1 ]
IDs=[r'Sieve analysis',r'Hydrometer']
GF.PlotAll(GSD, IDs, 'd', 'P', hold=1, xlabel=r'Particle diameter ($d$) [mm]',\
          ylabel=r'Percentage finer', xsize=7, ysize=4, lims=lims, linewith=2)
plt.xscale('log')
plt.grid(True)
plt.grid(b=True, which='minor', linestyle='--')
plt.savefig('combinedGSD.svg')
```
