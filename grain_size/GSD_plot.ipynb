```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
size=np.array([0.075,0.15,0.425,2,4.75,9.5,19,38,75])
PP=np.array([4,5,8,20,27,36,49,70,100])
label=['3 in', '1.5 in', '3/4 in', '3/8 in', 'No. 4', 'No. 10', 'No. 40', 'No. 100', 'No. 200']
#invert label order
label=label[::-1]
```

```python
import matplotlib.pyplot as plt

fig, ax_left = plt.subplots()

# Semilog X
ax_left.set_xscale('log')
ax_left.set_xlim(1e-2, 1e2)

# Left Y: 0% to 100%
ax_left.set_ylim(0, 100)
ax_left.set_ylabel('Percent Passing or finer than (%)')

# Right Y: 100% to 0% (mirrored)
ax_right = ax_left.twinx()
ax_right.set_ylim(100, 0)
ax_right.set_ylabel('Percent Retained or Coarser than (%)')

ax_left.set_xlabel('Grain size (mm)')
ax_left.grid(True, which='both')

#Flip x axis
ax_left.invert_xaxis()

#Make plot be 8 in wide by 4 in tall
fig.set_size_inches(8, 4.5)

#Plot the sieve size as vertical lines for reference
for s, l in zip(size, label):
    ax_left.axvline(x=s, color='red', linestyle='--', linewidth=0.7)
    ax_left.text(s, 102, l, rotation=90, verticalalignment='bottom', horizontalalignment='center', fontsize=8)

#Set vertical ticks every 10%
ax_left.set_yticks(np.arange(0, 110, 10))
ax_right.set_yticks(np.arange(0, 110, 10))


#plot the data over
ax_left.plot(size, PP, marker='o', color='blue', linestyle='-')
plt.tight_layout()
#save as svg
plt.savefig('GSD_plot.svg')
```
